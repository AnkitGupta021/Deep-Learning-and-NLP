```python
## Weight Initiailization
## There are 2 techniques
## 1.Xavier/Glorot Initialization for Normal and Uniform distribution
## 2. He Initialization for Normal and Uniform Distribution
## In keras there is parameter known as kerel_initializer which can be given 4 inputs,
## 'glorot_uniform' or 'glorot_normal'
## 'he_uniform' or 'he_normal

## Note :  glorot should always be used with 'tahn' and he should always be used with 'relu'
## Default value of kernel_initializer is 'glorat_uniform'
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import make_moons


import tensorflow as tf
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
```

```python
## Generating Binary Classification Dataset

x,y = make_moons(n_samples=250, noise=0.25, random_state=1)
```

```python
x
```

```python
y
```

```python
## Data Visualization

plt.scatter(x[:,0], x[:,1],c=y)
```

# Glorot Intialization for Normal distribution with 'tanh' as an Activation function

```python
## Building ANN

model = Sequential()
model.add(Dense(10, activation='tanh', input_dim=2, kernel_initializer='glorot_normal'))
model.add(Dense(10, activation='tanh',kernel_initializer='glorot_normal'))
model.add(Dense(10, activation='tanh',kernel_initializer='glorot_normal'))
model.add(Dense(10, activation='tanh',kernel_initializer='glorot_normal'))
model.add(Dense(1, activation='sigmoid'))
model.summary()
```

```python
## Training the Model

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
history = model.fit(x,y, epochs= 100, validation_split = 0.2 , verbose =0  )
```

```python
# Checking the decsion boundaries

from mlxtend.plotting import plot_decision_regions

plot_decision_regions(x,y, clf =model, legend =2)
```

 Model is showing better result.

# Glorot Intialization for Uniform distribution with 'tanh' as an Activation function

```python
## Buulding ANN

model = Sequential()
model.add(Dense(10, activation='tanh', input_dim=2, kernel_initializer='glorot_uniform'))
model.add(Dense(10, activation='tanh',kernel_initializer='glorot_uniform'))
model.add(Dense(10, activation='tanh',kernel_initializer='glorot_uniform'))
model.add(Dense(10, activation='tanh',kernel_initializer='glorot_uniform'))
model.add(Dense(1, activation='sigmoid'))
model.summary()
```

```python
## Training the Model

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
history = model.fit(x,y, epochs= 100, validation_split = 0.2, verbose=0 )
```

```python
# Checking the decsion boundaries

from mlxtend.plotting import plot_decision_regions

plot_decision_regions(x,y, clf =model, legend =2)
```

Model showing better results that glorot normal

# He Intialization for Normal distribution with 'relu' as an Activation function

```python
## Buulding ANN

model = Sequential()
model.add(Dense(10, activation='relu', input_dim=2 , kernel_initializer='he_normal'))
model.add(Dense(10, activation='relu', kernel_initializer='he_normal'))
model.add(Dense(10, activation='relu', kernel_initializer='he_normal'))
model.add(Dense(10, activation='relu', kernel_initializer='he_normal'))
model.add(Dense(1, activation='sigmoid'))
model.summary()
```

```python
## Training the Model

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
history = model.fit(x,y, epochs= 100, validation_split = 0.2, verbose=0 )
```

```python
# Checking the decision boundaries

from mlxtend.plotting import plot_decision_regions

plot_decision_regions(x,y, clf =model, legend =2)
```

Here the results are better than glorot with 'tanh' in both cases
