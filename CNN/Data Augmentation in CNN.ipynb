```python
# Importing the Libraries

import tensorflow as tf
from tensorflow import keras
from keras import Sequential
from keras.preprocessing.image import ImageDataGenerator
from keras.preprocessing import image

```

```python
# Loading the Image
# To load the image we will be using the image.load_img() where we pass the path and image size


img = image.load_img('Catu.jpeg', target_size=(200,200))
```

```python
# Checking the Result

img
```

```python
# Checking the Type of Image
type(img)
```

```python
# Connverting Img to array

## Since the type of image is Not Array its PIL object we need to convert it into array first to perform the 
## Transformation and we will be using image.img_to_array() function 

img_arry = image.img_to_array(img)

#Checking the result

type(img)
```

## Performing Tranformation on Image

###  Rotation

```python
## We will first create an object of ImageDataGenerator and give some value to it rotation argument
## Then we use ImageDataGenerator.random_transform() to apply the changes, which takes orginal image array and
## return the transformed image in array format. If we want to fix that the result can assign random value to 
## 'seed' as 2nd argument. By Default, 'seed'=None
## To get the final transform Image we will be using image.array_to_image(), where we pass transform array
## and it return us the image and then we compare both Original and Transformed Image



obj = ImageDataGenerator(rotation_range=40)
trans1_arry = obj.random_transform(img_arry)
trans1_img = image.array_to_img(trans1_arry)

# Comaparing the image 

plt.figure(figsize=(10,8))

plt.subplot(1,2,1)
plt.title('Original Image')
plt.imshow(img)

plt.subplot(1,2,2)
plt.title('Transformed Image')
plt.imshow(trans1_img)

plt.show()
```

### Horizontal Flipping 

```python
# Horizontal Flip

obj = ImageDataGenerator(horizontal_flip=True)
trans1_arry = obj.random_transform(img_arry, seed=1)
trans1_img = image.array_to_img(trans1_arry)

# Comaparing the image 

plt.figure(figsize=(10,8))

plt.subplot(1,2,1)
plt.title('Original Image')
plt.imshow(img)

plt.subplot(1,2,2)
plt.title('Transformed Image')
plt.imshow(trans1_img)

plt.show()

```

### Vertical Flip

```python
# Vertical Flip

obj = ImageDataGenerator(vertical_flip=True)
trans1_arry = obj.random_transform(img_arry)
trans1_img = image.array_to_img(trans1_arry)

# Comaparing the image 

plt.figure(figsize=(10,8))

plt.subplot(1,2,1)
plt.title('Original Image')
plt.imshow(img)

plt.subplot(1,2,2)
plt.title('Transformed Image')
plt.imshow(trans1_img)

plt.show()
```

## Shearing

```python
# shear_range >> Float. Shear Intensity (Shear angle in counter-clockwise(anti-clockwise) direction in degrees)


obj = ImageDataGenerator(shear_range=30)
trans1_arry = obj.random_transform(img_arry)
trans1_img = image.array_to_img(trans1_arry)

# Comparing the image 

plt.figure(figsize=(10,8))

plt.subplot(1,2,1)
plt.title('Original Image')
plt.imshow(img)

plt.subplot(1,2,2)
plt.title('Transformed Image')
plt.imshow(trans1_img)

plt.show()
```

## Shifting the Image

```python
## Height shifting 


obj = ImageDataGenerator(height_shift_range=0.9)
trans1_arry = obj.random_transform(img_arry)
trans1_img = image.array_to_img(trans1_arry)

# Comparing the image 

plt.figure(figsize=(10,8))

plt.subplot(1,2,1)
plt.title('Original Image')
plt.imshow(img)

plt.subplot(1,2,2)
plt.title('Transformed Image')
plt.imshow(trans1_img)

plt.show()
```

### Zooming 

```python
# zoom_range > Float or [lower, upper]. Range for random zoom. 
# If a float, [lower, upper] = [1-zoom_range, 1+zoom_range].


obj = ImageDataGenerator(zoom_range=(0.7,0.8))
trans1_arry = obj.random_transform(img_arry)
trans1_img = image.array_to_img(trans1_arry)

# Comparing the image 

plt.figure(figsize=(10,8))

plt.subplot(1,2,1)
plt.title('Original Image')
plt.imshow(img)

plt.subplot(1,2,2)
plt.title('Transformed Image')
plt.imshow(trans1_img)

plt.show()

```

### More about fill_mode argument

```python
# More About fill_mode = , it has 4 values, 


# One of {"constant", "nearest", "reflect" or "wrap"}. Default is 'nearest'. Points outside the boundaries 
# of the input are filled according to the given mode:
# 'constant': kkkkkkkk|abcd|kkkkkkkk (cval=k)
# 'nearest': aaaaaaaa|abcd|dddddddd
# 'reflect': abcddcba|abcd|dcbaabcd
# 'wrap': abcdabcd|abcd|abcdabcd 


## Height shifting with Fill_mode= 'reflect' , it reflect the transformed Part


obj = ImageDataGenerator(height_shift_range=0.9, fill_mode='reflect')
trans1_arry = obj.random_transform(img_arry)
trans1_img = image.array_to_img(trans1_arry)

# Comparing the image 

plt.figure(figsize=(10,8))

plt.subplot(1,2,1)
plt.title('Original Image')
plt.imshow(img)

plt.subplot(1,2,2)
plt.title('Transformed Image')
plt.imshow(trans1_img)

plt.show()
```

```python
## Height shifting with Fill_mode= 'nearest' , it fills the shifted region with nearest pixels

obj = ImageDataGenerator(height_shift_range=0.9, fill_mode='nearest')
trans1_arry = obj.random_transform(img_arry)
trans1_img = image.array_to_img(trans1_arry)

# Comparing the image 

plt.figure(figsize=(10,8))

plt.subplot(1,2,1)
plt.title('Original Image')
plt.imshow(img)

plt.subplot(1,2,2)
plt.title('Transformed Image')
plt.imshow(trans1_img)

plt.show()
```

```python
## Height shifting with Fill_mode= 'constant' , it fill the tranform region with 'Black' color


obj = ImageDataGenerator(height_shift_range=0.9, fill_mode='constant')
trans1_arry = obj.random_transform(img_arry)
trans1_img = image.array_to_img(trans1_arry)

# Comparing the image 

plt.figure(figsize=(10,8))

plt.subplot(1,2,1)
plt.title('Original Image')
plt.imshow(img)

plt.subplot(1,2,2)
plt.title('Transformed Image')
plt.imshow(trans1_img)

plt.show()
```

```python
## Height shifting with Fill_mode= 'wrap' , it creates the boundary. 

obj = ImageDataGenerator(height_shift_range=0.9, fill_mode='wrap')
trans1_arry = obj.random_transform(img_arry)
trans1_img = image.array_to_img(trans1_arry)

# Comparing the image 

plt.figure(figsize=(10,8))

plt.subplot(1,2,1)
plt.title('Original Image')
plt.imshow(img)

plt.subplot(1,2,2)
plt.title('Transformed Image')
plt.imshow(trans1_img)

plt.show()
```

## Generating Multiple Images from Exisiting Image

```python
## For this we will use the ImageDataGenerator.flow() function 


# flow(
#     x,
#     y=None,
#     batch_size=32,
#     shuffle=True,
#     sample_weight=None,
#     seed=None,
#     save_to_dir=None,
#     save_prefix='',
#     save_format='png',
#     ignore_class_split=False,
#     subset=None
# )

# Usefull parameters will be
# x = np.array of the image which we want to create the copies
# batch_size = size of batch, In case of single image we will give 1, default = 32
# seed = If we want to fix the randomization, we assign a value to it
# save_to_dir = directory where we want to save our images
# save_prefix = If we want to give any prefix to the Image Name of transformed Images for identifcation purpose.
# save_format = format in which we want the images to be saved, by default it 'png'


# Steps we follow are:

# > Create a ImageDataGenerator() object
# > Convert the Image into Tensor, or batch
# > Run a loop upto no.of Images we needed

# Step1 : Create ImageDataGenerator() object

obj1 = ImageDataGenerator(
    rotation_range=30,
    shear_range=0.2,
    width_shift_range= 0.2,
    height_shift_range= 0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='reflect'
    )
```

```python
## Step 2: Converting the Image into Batch/tensor
## As when we pass any image we need to pass it in batches, so here we will convert our 3D into 4D.

img_arry.ndim
```

```python
# Converting into Tensor

img = img_arry.reshape((1,200,200,3))
img.ndim
```

```python
## Step 3:  Running Inifinite Loop and break it utpill No. of Transformed Images we needed

## For 10 Images

i =0

for o in obj1.flow(img, batch_size=1, save_to_dir='Oct'):
    i+=1
    
    if i==10:
        break
        
## If we check out Directory 'Oct' there should be 10 images stored there. 
```
