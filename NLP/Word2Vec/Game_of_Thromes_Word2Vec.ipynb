Word2vec is a technique for natural language processing (NLP) published in 2013. The word2vec algorithm uses a neural network model to learn word associations from a large corpus of text. Once trained, such a model can detect synonymous words or suggest additional words for a partial sentence.

In this Code,we will be learning how to create the Word2Vec Model
Dataset we are going to use is The Games of Thromes Books dataset from kaggle and the library we are going to use is gensim library to create a word2vec model.
The idea is 
- collate the data of all the books to found a corpus
- Perform basis text preprocessing
- Build a model
- Try to do visualizaition

```python
# Importing other dependencies


import numpy as np
import pandas as pd
import gensim
import os    # to extract data from directory 'data'

from nltk import sent_tokenize # for sentence tokenization
from gensim.utils import simple_preprocess # for simple preprocessing
```

```python
# Data Gathering 
# Idea is to loop over all the files present in directory 'data' collate, perform sentence tokenization, and basic
# text preprocessing and then store it in a list. (Corpus)

story = []

# Function to remove stopwords:

def remove_stopwords(text):
    new_text = []
    
    for word in text.split():
        if word in stopwords.words('english'):
            new_text.append('')
        else:
            new_text.append(word)
    return " ".join(new_text)

for filename in os.listdir('Data'): 
    
    f = open(os.path.join('Data', filename), encoding='cp1252') # File Handler
    corpus = f.read()
    c = remove_stopwords(corpus)
    raw_data = sent_tokenize(c) # To perform sentence tokenization
    
    #Perform basis preprocessing and then store it in story
    for data in raw_data:   
        story.append(simple_preprocess(data))
        
```

```python
## Building the Model
# Now to build the model we will using gemsin library
# In gensim library we have model as gensim.models.Word2Vec() to define the model.
# We will define follow parameters in that 
# window - Maximum distance between the current and predicted word within a sentence.In denotes type of no. of 
# context on before and after the target word. window =1, set of 3 word, context<> Target<>context

# min_count = Ignores all words with total frequency lower than this.
# vector_size = Dimensionality of the word vectors, default is 100
# epochs = Number of iterations (epochs) over the corpus. default=5

model = gensim.models.Word2Vec(
    window = 10, 
    min_count = 2,
    vector_size= 150) # we will keep dimension as 150
```

```python
# After the model is created we will build out vocabulary from the corpus using atribute, .build_vocab()

model.build_vocab(story)
```

```python
model.corpus_count
```

```python
# Training the Model > Model will be trained using model.train()
# We will pass, 
# corpus data = Story
#total_examples = total # of words in corpus, we can be obtained using model.corpus_count
# epochs = model.epochs (by default=5) we can also give our own value.

model.train(story, total_examples=model.corpus_count,epochs= model.epochs)
```

### Visualization

```python
# Since we out vectors are in 150-D we need to reduce the dimension to visualize them
# We will use PCA as dimention reduction technique to visualize the 3D plot for first few hundred vecotrs

from sklearn.decomposition import PCA
```

```python
# Initializing the PCA object

pca = PCA(n_components=3)
```

```python
# Applying PCA on normed_vector and store it in a variable

X = pca.fit_transform(model.wv.get_normed_vectors())
```

```python
# Checking the shape

X.shape
```

```python
# Ploting 
# We will be using plotly.express library to plot the 3D scatter plot for 1st 100 words 

import plotly.express as px
fig = px.scatter_3d(X[200:300],x=0,y=1,z=2, color=y[200:300])
fig.show()
```
