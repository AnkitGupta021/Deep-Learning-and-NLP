```python
# Importing the dependencies

import pandas
import requests
```

```python
# Getting a response variable consisting all the data

response = requests.get("https://api.themoviedb.org/3/movie/top_rated?api_key=63a0965b07eac923c5a327e98e4e77fc&lenaguage=en-US&page=1")
```

```python
# response.json() will give all the data and response.json['results'] will give the keys values associated with 
# 'results' keys

response.json()['results']
```

```python
# Creating a Dataframe with selective columns

df = pd.DataFrame(response.json()['results'])[['id','title', 'overview','release_date', 'popularity', 'vote_average','vote_count']]
```

```python
# creating a Dataframe to get all the rows for selective columns 

df1 = pd.DataFrame() # empty datafrme
```

```python
# Using for loop extracting the data from API link
# we have total 449 pages so we willl choose range(1,450)
# Use formatting to count from each pages
# Store the value of each page in temp_dp and then concatanate with main df1

for i in range(1,450):
    response = requests.get(f"https://api.themoviedb.org/3/movie/top_rated?api_key=63a0965b07eac923c5a327e98e4e77fc&lenaguage=en-US&page={i}")
    temp_df = pd.DataFrame(response.json()['results'])[['id','title', 'overview','release_date', 'popularity', 'vote_average','vote_count']]
    df1 = pd.concat([df1,temp_df], ignore_index=True)
```
